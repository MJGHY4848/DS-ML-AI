#### The concept of missing values refers to non availability of data for certain columns or observations we are obtaining data for. this may occur for various reasons like data was not collected or error occured during data entry or some other reaons as well.

#### We can use following methods to identify missing values in a dataframe, such as:
#### i) isna() and isnull(): These methods return a DataFrame where each cell is True if the corresponding element is NaN or None, and False otherwise.
#### ii) notna() and notnull(): These are the methods that return True for non-missing values and False for missing values.

#### Different Ways of Treating/Imputing/Deleting Missing Values:

#### Imputaion means to fill missing values with estimated values based on the remaining data of the same column. it involves mean, median, mode or forward filling, backward filling and some othr methods as well
#### Example of Imputation:

```python
import pandas as pd
import numpy as np

data = {
    'A': [1, 2, np.nan, 4],
    'B': [5, np.nan, np.nan, 8],
    'C': [np.nan, 10, 11, 12]
}

df = pd.DataFrame(data)
df
```

```python
df.isnull()
```

```python
df.fillna(df.mean())
```

#### Deleting missing values means to delete the rows which contains NaN values, it can be both ,either row deletion or column deletion

```python
df.dropna(axis=0, inplace=True)
df
```
