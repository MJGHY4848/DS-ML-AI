# 3) Create 2 new columns based on Player name. First column will have first name and second column will have last name. Eg: for the player Shikhar Dhawan, Shikhar will be the first name and Dhawan will be the last name.

# 10) Players who scored atleast one century in this season. Create visualization.

# 19) Create visualization of top 10 players with highest number of sixes.

```python
highest_sixes = Ipl_data.sort_values("6s", ascending= False).head(10)
highest_sixes
```

```python
col=['b','g','c','m','y']
plt.xlabel('Players')
plt.ylabel('No. of Sixes')
plt.xticks(rotation=45, ha='right')
plt.title('Top Sixers',fontsize=20)
plt.bar(highest_sixes["Player"],highest_sixes["6s"],color=col)
for x,y in zip(highest_sixes["Player"],highest_sixes["6s"]):
    plt.text(x,y,y) 
plt.show()
```

# 20) Scatter plot of runs scored by a player v/s balls faced in a season. Then find the relationship between these 2 variables.
