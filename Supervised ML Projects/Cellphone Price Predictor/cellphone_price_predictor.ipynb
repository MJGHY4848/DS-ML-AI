# A.	Explain the main assumptions of Linear Regression in detail.

Linear regression is a powerful statistical method used to model the relationship between a dependent variable and one or more independent variables. However, for the results of a linear regression analysis to be valid, several key assumptions must be met. Here’s a detailed explanation of these assumptions:

1. Linearity: 
The relationship between the independent and dependent variables should be linear. This means that a change in the independent variable(s) should produce a proportional change in the dependent variable. This assumption can be checked by examining scatter plots of the independent variables against the dependent variable.

2. Independence: 
The residuals (the differences between observed and predicted values) should be independent. This means that the value of the residual for one observation should not be influenced by the value of the residual for another observation.

3. Homoscedasticity: 
The variance of the residuals should be constant across all levels of the independent variables. In other words, the spread of the residuals should remain the same regardless of the value of the independent variable(s). 

4. Normality of Residuals: 
The residuals should be approximately normally distributed, especially for the purposes of hypothesis testing and creating confidence intervals. This assumption can be checked using a histogram of the residuals or a Q-Q plot. If the residuals are not normally distributed, it may affect the validity of statistical tests related to the model.

5. No Multicollinearity: 
In multiple linear regression, the independent variables should not be too highly correlated with each other. Multicollinearity can inflate the variance of the coefficient estimates and make them unstable, which can affect the interpretation of the model. This can be assessed using Variance Inflation Factor (VIF) scores or correlation matrices.

6. No Significant Outliers: 
Outliers can have a disproportionate effect on the results of a linear regression analysis. It's important to check for outliers and influential points (using methods such as Cook's distance) as they can skew the results and lead to incorrect conclusions.

# B.What is the difference between R-squared and Adjusted R-squared?

R-squared and Adjusted R-squared are both statistical metrics used to evaluate the goodness of fit of a regression model.

R-Squared
Definition: R-squared measures the proportion of the variance in the dependent variable that can be explained by the independent variable(s) in the model. It provides a value between 0 and 1, where:

0 indicates that the model does not explain any of the variance.
1 indicates that the model explains all the variance.

Increases with More Predictors: R-squared will always increase (or at least not decrease) when additional independent variables are added to the model, regardless of whether these variables are actually meaningful or not. This can lead to overfitting.

Adjusted R-squared
Definition: Adjusted R-squared adjusts the R-squared value based on the number of predictors in the model. It accounts for the number of independent variables and the sample size, providing a more accurate measure of model fit, especially when comparing models with different numbers of predictors.

Advantages:
Penalizes Overfitting: Adjusted R-squared can decrease if unnecessary predictors are added, making it a better metric for model comparison.
More Reliable: It provides a more reliable indication of the goodness of fit, especially when comparing models with different numbers of predictors.

# C.	What are the different types of Regularization techniques in Regression. Explain in detail with cost functions of each technique.

The two most common types of regularization techniques are Lasso Regression (L1 Regularization) and Ridge Regression (L2 Regularization). There's also Elastic Net, which combines both techniques. 

1. Ridge Regression (L2 Regularization)
Concept: Ridge regression adds a penalty equivalent to the square of the magnitude of coefficients. This helps to reduce the complexity of the model by constraining the coefficients, especially in the presence of multicollinearity.

2. Lasso Regression (L1 Regularization)
Concept: Lasso regression adds a penalty equal to the absolute value of the magnitude of coefficients. This can lead to some coefficients being exactly zero, effectively performing variable selection.

3. Elastic Net Regularization
Concept: Elastic Net is a hybrid of Lasso and Ridge regression, combining both L1 and L2 penalties. This approach is useful when there are multiple features that are correlated with each other.



# D.	How logistic regression works for multiclass classification. Explain in detail.

In multiclass classification, where the target variable can belong to one of several classes (say K classes), the approach is slightly different.

a. One-vs-Rest (OvR) Method
In the One-vs-Rest approach, you train K separate binary classifiers, one for each class. Each classifier predicts the probability that a sample belongs to its corresponding class against all other classes combined.
After training K classifiers, to predict the class of a new instance, you compute the probabilities for each class and select the one with the highest probability

b. Softmax Regression
Softmax regression is a generalization of logistic regression that directly computes the probabilities for all classes in one model. Instead of estimating probabilities for each class separately, it uses the softmax function to output a probability distribution over K classes.
The softmax function transforms the raw scores (logits) from the linear combinations of features into probabilities.

# E.	Explain the performance metrics of logistic regression.

Evaluating the performance of logistic regression, especially in classification tasks, involves several key metrics. These metrics help to assess how well the model is predicting outcomes. Here’s a detailed overview of the primary performance metrics used for logistic regression:

1. Confusion Matrix
A confusion matrix is a table that summarizes the performance of a classification model. For binary classification, it has four components:

True Positives (TP): Correctly predicted positive cases.
True Negatives (TN): Correctly predicted negative cases.
False Positives (FP): Incorrectly predicted positive cases (Type I error).
False Negatives (FN): Incorrectly predicted negative cases (Type II error).

2. Accuracy
Accuracy is the ratio of correctly predicted instances (both true positives and true negatives) to the total number of instances.

3. Precision
Precision (also called Positive Predictive Value) measures the accuracy of positive predictions. 

4. Recall (Sensitivity or True Positive Rate)
Recall measures the ability of the model to identify all relevant instances (true positives).

5. F1 Score
The F1 Score is the harmonic mean of precision and recall, providing a balance between the two. It is particularly useful when you need to find an optimal balance between precision and recall.

6. Receiver Operating Characteristic (ROC) Curve
The ROC curve is a graphical representation of the trade-off between sensitivity (true positive rate) and specificity (1 - false positive rate) across different thresholds.

7. Area Under the ROC Curve (AUC-ROC)
The AUC-ROC quantifies the overall performance of the model across all thresholds. It represents the probability that a randomly chosen positive instance is ranked higher than a randomly chosen negative instance.
